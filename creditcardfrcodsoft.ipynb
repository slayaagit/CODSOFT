```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score, average_precision_score, precision_recall_curve, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('creditcard.csv')
data
```

```python
data.isnull().sum()
```

```python
X=data.drop("Class",axis=1)
y=data.Class
```

```python
st=StandardScaler()
X=st.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model=LogisticRegression(penalty="l2",solver="lbfgs",max_iter=100)
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
y_predictp = model.predict_proba(X_test)[:, 1]
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_predict), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
precision,recall,thresholds=precision_recall_curve(y_test,y_predictp)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label='Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
f1_scores=2*(precision*recall)/(precision+recall)
best_f1_score=np.max(f1_scores)
best_threshold=thresholds[np.argmax(f1_scores)]
print("f1_scores=" ,best_f1_score)
print("best thr=" ,best_threshold)
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_predict))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_predict))
print("Accuracy Score =", accuracy_score(y_test, y_predict))
print("ROC AUC Score =", roc_auc_score(y_test, y_predictp))
```
